```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from bertopic import BERTopic
from umap import UMAP
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from hdbscan import HDBSCAN
from nltk.corpus import stopwords
from transformers.pipelines import pipeline

from flair.embeddings import TransformerDocumentEmbeddings

import re

from wordcloud import WordCloud
```

–ù—É–∂–Ω–æ –ø–æ—á–∏—Å—Ç–∏—Ç—å —Ç–µ–∫—Å—Ç—ã

```python
def clean_text(text):
    #import re –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ pandarallel
    try:
        text = text.lower()
        text = re.sub('[^\w\s]', ' ', text) # —É–¥–∞–ª—è–µ–º –≤—Å–µ —Å–∏–º–≤–æ–ª—ã, –∫—Ä–æ–º–µ –±—É–∫–≤ –∏ —Ü–∏—Ñ—Ä
        text = re.sub('\d+', '', text) # —É–¥–∞–ª—è–µ–º –≤—Å–µ —Ü–∏—Ñ—Ä—ã
        text = re.sub('[^–∞-—è–ê-–Ø\s]+' , '', text)
        text = re.sub('\s+', ' ', text) # –∑–∞–º–µ–Ω—è–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –ø—Ä–æ–±–µ–ª–æ–≤ –æ–¥–Ω–∏–º –ø—Ä–æ–±–µ–ª–æ–º
        
        text = text.strip() # —É–¥–∞–ª—è–µ–º –ø—Ä–æ–±–µ–ª—ã –≤ –Ω–∞—á–∞–ª–µ –∏ –∫–æ–Ω—Ü–µ —Ç–µ–∫—Å—Ç–∞
    except:
        pass
    return text
```

```python
posts_df['post_text_cleaned'] = posts_df.post_text.apply(clean_text)
posts_df
```

–ß–∏—Å—Ç–∏–º —Ç–µ–∫—Å—Ç—ã –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ df

–°–ª–æ–≤–æ `—ç—Ç–æ` nltk –Ω–µ —Å—á–∏—Ç–∞–µ—Ç —Å—Ç–æ–ø-—Å–ª–æ–≤–æ–º, –∞ –º—ã –¥–æ–±–∞–≤–∏–º

```python
stop_words = stopwords.words('russian')
stop_words.extend(['—ç—Ç–æ','—Ç–∞','—ç—Ç–∞','–æ–±–æ','—Å–≤–æ–π'])
```

```python
roberta = TransformerDocumentEmbeddings('xlm-roberta-large')
topic_model = BERTopic(embedding_model=roberta)
```

–ê–ª–≥–æ—Ä–∏—Ç–º —Å–Ω–∏–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –∏–º–µ–µ—Ç —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ. –ü–æ—ç—Ç–æ–º—É –∑–∞—Ñ–∏–∫—Å–∏—Ä—É–µ–º –µ–≥–æ –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

```python
umap_model = UMAP(n_neighbors=15, 
                  n_components=5, 
                  min_dist=0.0, 
                  metric='cosine', 
                  random_state=42)
```

–ù–∞—Å—Ç—Ä–æ–∏–º –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é, –µ—Å—Ç—å 2 –ø—É—Ç–∏:
- —Å—Ç–∞–Ω–¥–∞—Ä–Ω—Ç—ã–π HDBSCAN, —Ö–æ—Ä–æ—à –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –ø–ª–æ—Ö —Ç–µ–º, —á—Ç–æ –æ—Ç–¥–∞–ª–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–Ω–∏–º–∞—é—Ç—Å—è –∑–∞ –≤—ã–±—Ä–æ—Å
- KMeans, —Ö–æ—Ä–æ—à —Ç–µ–º, –Ω–∞–∑–Ω–∞—á–∞–µ—Ç –æ—Ç–¥–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Ç–µ—Ä –≤—ã–±—Ä–æ—Å–∞–º, –ø–ª–æ—Ö –≤ –æ—Ç–Ω–æ—à–µ–Ω–∏–∏ —Ä–∞–±–æ—Ç—ã —Å –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏ —Ö–∏—Ç—Ä–æ–π —Ñ–æ—Ä–º—ã

–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã –æ–ø—Ä–µ–¥–µ–ª—è–ª–∏ —Å—Ö–æ–∂–∏–µ –∫–ª–∞—Å—Ç–µ—Ä—ã, –Ω–æ —Ç.–∫. KMeans –Ω–µ –¥–∞–µ—Ç –≤—ã–±—Ä–æ—Å–æ–≤ –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–º–µ–Ω–Ω–æ –µ–≥–æ.

–ü—Ä–æ–≤–µ—Ä—è—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–µ—Ç–æ–¥–æ–º –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø–ª–µ—á–∞ –Ω–µ –±—É–¥–µ–º, –∞ –ø—Ä–æ—Å—Ç–æ –≤–æ–∑—å–º–µ–º —Ç–æ–ø-20 —Ç–µ–º

```python
cluster_size = 15
```

```python
# cluster_model = KMeans(
#     n_clusters=25, 
#     random_state=42
# )

cluster_model = HDBSCAN(
    min_cluster_size=cluster_size, 
    metric='euclidean', 
    cluster_selection_method='eom', 
#     cluster_selection_epsilon = 0.3,
    prediction_data=True, 
    min_samples=2
)
```

–≤ –ø–æ—Å—Ç-–æ–±—Ä–∞–±–æ—Ç–∫–µ (–Ω–∞ –ø—Ä–µ–¥–ø–æ—Å–ª–µ–¥–Ω–µ–º —ç—Ç–∞–ø–µ) –±—É–¥–µ—Ç –æ—Ç—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å `CountVectorizer` —Å –Ω–∞—à–∏–º–∏ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞–º–∏.

`Tf-idf` –¥–ª—è –≤–∑–≤–µ—à–∏–≤–∞–Ω–∏—è –æ–∫–æ–Ω—á–∞—Ç–µ–ª—å–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —É–∂–µ –≤—à–∏—Ç –≤ `BERTopic`

```python
# vectorizer_model = CountVectorizer(stop_words=stopwords, min_df=1, max_df=0.6)
vectorizer_model = CountVectorizer(stop_words=stop_words)
```

–¢–µ–ø–µ—Ä—å —Å–æ–±–µ—Ä–µ–º –º–æ–¥–µ–ª—å.

–û–ø—Ä–µ–¥–µ–ª—è—Ç—å —Ç–µ–º—É –±—É–¥—É—Ç —Ç–æ–ø-10 —Å–ª–æ–≤

```python
topic_model = BERTopic(
    language="russian",
#     embedding_model=roberta,
    top_n_words=10,
    calculate_probabilities=True, 
    verbose=True, 
    vectorizer_model=vectorizer_model, 
    min_topic_size=cluster_size, 
    umap_model=umap_model, 
    hdbscan_model=cluster_model
)
```

```python
topics, probs = topic_model.fit_transform(concatenated_texts)
```

```python
def clean_string(string):
    result = []
    for i in range(len(string)):
#         –≤ —Å–ª—É—á–∞–µ –µ—Å–ª–∏ —Å–∏–º–≤–æ–ª –Ω–æ—Ä–º–∞–ª—å–Ω—ã–π, —Ç–æ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ result —Å–∏–º–≤–æ–ª
        if string[i] != '#':
            result += string[i]
#         –≤ –∏–Ω–æ–º —Å–ª—É—á–∞–µ —É–¥–∞–ª—è–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç –∏–∑ result
        else:
            result = result[:-1]
    result = ''.join(result)
    return result
            
```
