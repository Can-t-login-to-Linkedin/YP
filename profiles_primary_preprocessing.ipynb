```python
import pandas as pd
import numpy as np
```

```python
import json
```

объединяем 1

```python
developer_profiles = pd.concat([
    pd.read_json('datasets/total_profiles.json').T,
    pd.read_json('datasets/profiles_developer_kristina_86.json').T,
    pd.read_json('datasets/profiles_developer_senior.json').T
])
developer_profiles
```

объединяем 2

```python
developer_reactions = pd.concat([
    pd.read_json('datasets/total_reactions.json', orient='index'),
    pd.read_json('datasets/reactions_developer_kristina_86.json', orient='index'),
    pd.read_json('datasets/reactions_developer_senior.json', orient='index')
])
developer_reactions
```

Объединяем 3

```python
developer_posts = pd.concat([
    pd.read_json('datasets/total_posts.json', orient='index'),
    pd.read_json('datasets/posts_developer_kristina_86.json', orient='index'),
    pd.read_json('datasets/posts_developer_senior.json', orient='index')
])
developer_posts
```

Собираем последние 2 датафрейма в листы и джойним 

```python
developer_profiles['reactions'] = developer_reactions.apply(list, axis=1)
developer_profiles['posts'] = developer_posts.apply(list, axis=1)
developer_profiles
```

Удаляем задублировавшихся пользователей

```python
developer_profiles = developer_profiles[~developer_profiles.index.duplicated()]
```

2 дубликата по id убрали.

Видно, что некоторые данные (стаж) при сохранении результатов "поплыли". Колонки `company` и `employment_type` в некоторых случаях сменились местами. Посмотрим на эти профили

Судя по профилю участника `d-batmanov` эта информация о последнем месте работы. У остальных участников есть только общий опыт в колонке `experience`.

Жаль, что потеряли названия компаний. Заменим их на `unknown`. Колонку `employment_type` перезаполнять не будем

```python
developer_profiles.loc[developer_profiles['employment_type'].str.contains('yrs') | developer_profiles['employment_type'].str.contains('mos'), 'company'] = 'unknown'
developer_profiles
```

Все юзеры разработчики (developer). Релевантными будем считать юзеров, у которых общий опыт >= 3 лет, либо имеют `senior` в названии должности. `middle` в данном случае брать не будем, потому что в разных компаниях разные градации

```python
developer_profiles = developer_profiles[developer_profiles['experience'] >= 3 | developer_profiles.role.str.lower().str.contains('senior')]
developer_profiles.head()
```

Оставили 290 релевантных, сохраняем в json чтобы не терять структуру списка (иначе pandas превратит все в строку, из которой потом не вытащить отдельные тексты)
